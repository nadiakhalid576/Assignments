```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student_performance_dataset_cleaned.csv")

# Select numeric columns only
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
corr = numeric_df.corr()

# Show correlation of features with Final_Exam_Score
target_corr = corr["Final_Exam_Score"].sort_values(ascending=False)
print("Correlation of features with Final Exam Score:\n")
print(target_corr, "\n")

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Analysis")
plt.show()
```
