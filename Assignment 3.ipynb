```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("student_performance_dataset_cleaned.csv")

# ----------------------------
# 1. Bar Plot  Average Final Exam Score by Gender
# ----------------------------
plt.figure(figsize=(6,4))
sns.barplot(x="Gender", y="Final_Exam_Score", data=df, errorbar=None, palette="pastel")
plt.title("Average Final Exam Score by Gender")
plt.show()
print("Insight 1: One gender performs slightly better on average, but the gap is small.\n")

# ----------------------------
# 2. Histogram  Distribution of Study Hours
# ----------------------------
plt.figure(figsize=(6,4))
sns.histplot(df["Study_Hours_per_Week"], bins=15, kde=True, color="skyblue")
plt.title("Distribution of Study Hours per Week")
plt.xlabel("Study Hours per Week")
plt.ylabel("Number of Students")
plt.show()
print("Insight 2: Most students study 1020 hours weekly; very few study extremely low or high hours.\n")

# ----------------------------
# 3. Scatter Plot  Study Hours vs Final Exam Score
# ----------------------------
plt.figure(figsize=(6,4))
sns.scatterplot(x="Study_Hours_per_Week", y="Final_Exam_Score", hue="Pass_Fail", data=df, palette="Set2")
plt.title("Study Hours vs Final Exam Score")
plt.xlabel("Study Hours per Week")
plt.ylabel("Final Exam Score")
plt.show()
print("Insight 3: Students who study more hours generally score higher, but effort does not always guarantee success.\n")

# ----------------------------
# 4. Box Plot  Attendance Rate by Pass/Fail
# ----------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="Pass_Fail", y="Attendance_Rate", data=df, palette="coolwarm")
plt.title("Attendance Rate by Pass/Fail")
plt.show()
print("Insight 4: Passed students usually have higher attendance compared to failed students.\n")

# ----------------------------
# 5. Heatmap  Correlation of Numeric Features
# ----------------------------
plt.figure(figsize=(8,6))
corr = df[["Study_Hours_per_Week","Final_Exam_Score","Past_Exam_Scores","Attendance_Rate"]].corr()
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f")
plt.title("Correlation Between Numeric Features")
plt.show()
print("Insight 5: Final exam scores strongly correlate with past scores, study hours, and attendance, highlighting key success factors.\n")
```
